```python
# Add labels, a legend, and change the size of the heatmap

f = plt.figure(figsize=(6, 6)) # figure size 
plt.matshow(sub.corr(), fignum=f.number) # type of plot
plt.xticks(range(sub.shape[1]), sub.columns, fontsize=12, rotation=90) # x axis labels
plt.yticks(range(sub.shape[1]), sub.columns, fontsize=12) # y axis labels
cb = plt.colorbar() # add a colour legend (called colorbar)
cb.ax.tick_params(labelsize=12) # add font size
plt.title('Correlation Matrix', fontsize=14) # add title
```

```python
# Create a subplot with matplotlib
f,ax = plt.subplots(figsize=(8,8))

# Create the correlation heatmap in seaborn by applying a heatmap onto the correlation matrix and the subplots defined above.
corr = sns.heatmap(sub.corr(), annot = True, ax = ax) # The `annot` argument allows the plot to 
#place the correlation coefficients onto the heatmap.
```

### Q3: Discuss heatmap results

Comment: 
A correlation coefficient close to 1 indicates a strong positive relationship.
Therefore, the following variables are strongly correlated to happiness: "gdp_per_capita" (0.72), social_support(0.65), "healthy_life_expectancy" (0.68). Surprisingly, "freedom_to_make_life_choices" (0.57) and "perceptions_of_corruption" (0.42) are less strongly associated and therefore less important for determining happiness.

(Note: for "perceptions_of_corruption", the higher the score, the lower the "perceptions_of_corruption". So less corrrupt countries have scores closer to 1, and more corrupt countries have scores closer to 0.)

However, the big surprise is that "generosity" (0.082) has a correlation coefficient close to 0 and is therefore not correlated to happiness. This could mean that people in politically stable countries with higher gdp assume that their government will take care of donations. Supporting this possibility, generosity is weakly positively correlated with 2 variables, "freedom to make life choices" (0.16) and "perceptions of corruption" (0.22), suggesting that people donate more in corrupt countries because they don't trust the government to do it.

### Q4: Discuss scatterplot results

Comment: There is a very clear trend that is apparent on both plots, and that is that if happiness is defined as having a happiness score of 6 and over, there is no happiness possible for countries with low life expectancy or low gdp per capita. In other words, having high life expectancy (> 0.4) and having high gdp per capita (> 0.6) are necessary to have happiness.

## Categorical Plots

```python
# make a histogram of "perceptions of corruption"
# The argument "kde" add a line that encompasses the distribution

sns.histplot(df['perceptions_of_corruption'], bins = 20, kde = True)
```

```python
# make a category to hold values of "perceptions_of_corruption" that are under 0.2 (new column == "corruption_status")
df.loc[df['perceptions_of_corruption'] < 0.2, 'corruption_status'] = 'higher corruption'
```

```python
# make a category to hold values of social_support that are over 0.75 (new column == "")
df.loc[df['perceptions_of_corruption'] >= 0.2, 'corruption_status'] = 'lower corruption'
```

```python
# check the distribution
df['corruption_status'].value_counts(dropna = False)
```

```python
# check the new column
df.head()
```

```python
# Create a categorical plot in seaborn using the social support categories created above

sns.set(style="ticks")
g = sns.catplot(x="happiness_score", y="region", hue="corruption_status", data=df, aspect=2)
```

```python
# fig = g.get_figure()
g.savefig("out2.png")
```

### Q6: Discuss categorical plot results

Comment: I made a categorical value for political corruption called "corruption_status" and flags for "lower corruption" (over 0.2) and "higher corruption" (under 0.2). Surprisingly, although corruption is lower in the happiest countries in Western Europe or North America/ANZ, it does not seem to have any impact on happiness in other regions.
